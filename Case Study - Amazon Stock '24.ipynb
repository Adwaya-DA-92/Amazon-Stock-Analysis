# Amazon stock data analysis:

#### The below data pulled from (https://www.nasdaq.com/market-activity/stocks/amzn/historical) and reflected from June'24 - December'24.

```python
import numpy as np
import matplotlib.pyplot as plt

stock_file = r'D:\Excel_file\New folder\amzn-stock-2024.csv'
stock_symbol = 'AMZN'
high_value = 150.00
minimal_opening_price = 100.00
```

```python
# Determine the range of stock prices within the data

stock_opening = np.loadtxt(stock_file, skiprows=1, usecols=3, delimiter=',')

print(f'The lowest opening price for the {stock_symbol} stock was ${np.min(stock_opening):.2f}.')
print(f'The highest opening price for the {stock_symbol} stock was ${np.max(stock_opening):.2f}.')
```

```python
# Determine the number of high trading days

stock_highs = np.loadtxt(stock_file, skiprows=1, usecols=4, delimiter=',')

print(f'The number of days that saw a high for the {stock_symbol} stock that exceeded ${high_value:.2f} was {np.sum(stock_highs > high_value)}.')
```

```python
# Determine whether the stock had a low opening

opening_price_check = 'did' if (stock_opening < minimal_opening_price).any() else 'did not'

print(f'The {stock_symbol} stock {opening_price_check} have an opening price less than ${minimal_opening_price:.2f}.')
```

```python
# Determine the number and percent of days where closing lower than opening

stock_closing = np.loadtxt(stock_file, skiprows=1, usecols=1, delimiter=',')

print(f'The number of days of lower closing price than opening price observed for {stock_symbol} stock was {np.sum(stock_closing < stock_opening)}.')
print(f'The percent of days of lower closing price than opening price observed for {stock_symbol} stock was {np.sum(stock_closing < stock_opening)/len(stock_closing)*100:.2f}%.')
```

```python
# Determine the range of stock prices between high and low

stock_lows = np.loadtxt(stock_file, skiprows=1, usecols=5, delimiter=',')

print(f'The highest difference of {stock_symbol} stock prices between high and low was ${np.max(stock_highs-stock_lows):.2f}.')
print(f'The lowest difference of {stock_symbol} stock prices between high and low was ${np.min(stock_highs-stock_lows):.2f}.')
```

```python
# Visualize the plot of 'AMZN' stock's closing price trends by dates

x = np.arange(0, len(stock_closing))
y = stock_closing

plt.title('Amazon Stock Closing Price')
plt.xlabel('Day')
plt.ylabel('Price')
plt.plot(x, y, color='blue')
plt.show
```

###### The below findings from the 'AMZN' stock analysis:

> The range of opening stock price is around 150.00 and 230.00 respectively.

> The high trading day's were always excessed 150.00.

> The opening stock price has never gone below 100.00.

> The percentage of lower closing price days than opening price were 52.76 (67 days), which is a little threat of the compay.

> The range of difference of stock prices between 11.35 and 1.49 respectively.

> According to above visualization, the closing price has gradually downed with the aspects of dates.
